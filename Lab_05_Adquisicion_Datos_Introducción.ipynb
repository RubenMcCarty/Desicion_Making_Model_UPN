# AGREGAR CABEZERA

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
df.columns = headers
df.head()
```

```python
df.tail(30)
```

necesitamos reemplazar el "?" símbolo con NaN para que dropna () pueda eliminar los valores faltantes

```python
df1 = df.replace ('?', np.NaN)
```

```python
df1
```

Ahora, hemos leído correctamente el conjunto de datos sin procesar y agregamos los encabezados correctos en el marco de datos.

```python
df1.head(10)
```

Podemos eliminar los valores que faltan a lo largo de la columna "precio" de la siguiente manera

```python
df=df1.dropna(subset=["price"], axis=0)
df.head(20)
```

Ahora, hemos leído correctamente el conjunto de datos sin procesar y agregamos los encabezados correctos en el marco de datos.

# Conocimiento básico del conjunto de datos
Después de leer los datos en el DF de Pandas, es hora de que exploremos el conjunto de datos.
Hay varias formas de obtener información esencial de los datos para ayudarnos a comprender mejor nuestro conjunto de datos.

# Tipos de datos
Los datos tienen una variedad de tipos.
Los principales tipos almacenados en los DF de Pandas son object, float, int, bool y datetime64. Para conocer mejor cada atributo, siempre es bueno que sepamos el tipo de datos de cada columna. En Pandas:

```python
df.dtypes # me indica el tipo de datos de cada columna, devuelve una serie con el tipo de datos de cada columna.
```

devuelve una serie con el tipo de datos de cada columna.

```python
print(df.dtypes)
```

Como resultado, como se muestra arriba, es claro ver que el tipo de datos de "symboling" y "curb-wieght" son int64, "pérdidas normalizadas" es objeto, y "base de ruedas" es float64, etc.

Estos tipos de datos se pueden cambiar; aprenderemos cómo lograr esto en un módulo posterior.

# Describe
Si quisiéramos obtener un resumen estadístico de cada columna, como el recuento, el valor medio de la columna, la desviación estándar de la columna, etc. Usamos el método de describe.

```python
df.describe()
```

Esto muestra el resumen estadístico de todas las columnas de tipo numérico (int, float).
Por ejemplo, el atributo "simbolización" tiene 205 recuentos, el valor medio de esta columna es 0,83, la desviación estándar es 1,25, el valor mínimo es -2, el percentil 25 es 0, el percentil 50 es 1, el percentil 75 es 2 y el valor máximo es 3.
Sin embargo, ¿qué pasaría si también quisiéramos comprobar todas las columnas, incluidas las que son de tipo objeto?

NOTA: Este método proporcionará varias estadísticas de resumen, excluyendo los valores de NaN (No es un número).

Puede agregar un argumento include = "all" dentro del corchete. Vamos a intentarlo otra vez.

```python
# Si desea leer todo el dato
df.describe(include = 'all')
```
